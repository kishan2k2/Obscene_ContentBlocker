```python
import csv

def calculate_accuracy(csv_file_path):
    # Initialize counters for same and different values
    same_count = 0
    total_count = 0

    # Open the CSV file
    with open(csv_file_path, 'r') as file:
        # Create a CSV reader
        csv_reader = csv.DictReader(file)

        # Iterate through each row in the CSV file
        for row in csv_reader:
            # Access the "Target" and "Prediction" values from the row
            target_value = row['Target']
            prediction_value = row['prediction']

            # Compare the values and update counters
            if target_value == prediction_value:
                same_count += 1

            total_count += 1

    # Calculate and return the accuracy
    accuracy = same_count / total_count if total_count > 0 else 0
    return accuracy

# Example usage
csv_file_path = 'csvfile.csv'
accuracy = calculate_accuracy(csv_file_path)
print(f'Accuracy: {accuracy:.2%}')
```
